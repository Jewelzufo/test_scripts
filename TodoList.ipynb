```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class TodoList:
    def __init__(self):
        self.todos = []
        self.create_widgets()
        self.display_ui()

    def create_widgets(self):
        self.input_text = widgets.Text(placeholder="Enter a new todo")
        self.add_button = widgets.Button(description="Add")
        self.add_button.on_click(self.add_todo)
        self.todo_list = widgets.VBox()
        self.output = widgets.Output()

    def display_ui(self):
        display(widgets.VBox([
            widgets.HTML("<h2>Todo List</h2>"),
            widgets.HBox([self.input_text, self.add_button]),
            self.todo_list,
            self.output
        ]))

    def add_todo(self, _):
        todo_text = self.input_text.value.strip()
        if todo_text:
            todo_item = widgets.Box([
                widgets.Label(todo_text),
                widgets.Button(description="Remove")
            ])
            todo_item.children[1].on_click(lambda _: self.remove_todo(todo_item))
            self.todos.append(todo_item)
            self.input_text.value = ""
            self.update_todo_list()

    def remove_todo(self, todo_item):
        self.todos.remove(todo_item)
        self.update_todo_list()

    def update_todo_list(self):
        with self.output:
            clear_output()
            self.todo_list.children = self.todos

# Create and display the TodoList
todo_list = TodoList()
```
